```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt


df = pd.read_excel('./Data/car_data_selection.xlsx')

cat_cols = ["symboling","carlength","aspiration","carbody","cylindernumber"]
df[cat_cols] = df[cat_cols].astype('category')

categ_variables = df[cat_cols]

#Create the contingency table
data_crosstab = pd.crosstab(categ_variables['symboling'],
                            categ_variables['carlength'], 
                               margins = False)

#Cramer V test
def cramers_corrected_stat(confusion_matrix):
    """ calculate Cramers V statistic for categorial-categorial association.
    """
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = sum(confusion_matrix.sum())
    return np.sqrt(chi2 / (n*(min(confusion_matrix.shape)-1)))

V = cramers_corrected_stat(data_crosstab)

association_level = ''
if(V<0.1):
    association_level='Little, if any association.'
else:
    if(V<0.3):
        association_level="Low association."
    else:
        if(V<0.5):
            association_level="Moderate association."
        else:
            association_level="High association"
print('Cramers V: ', V, '\n', association_level)

import association_metrics as am

#Cramer V matrix for all categorical variables
cramersv = am.CramersV(categ_variables)
cramersv.fit()




#
```
