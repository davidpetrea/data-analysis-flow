```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

df = pd.read_excel('./Data/car_data_selection.xlsx')

categ_names = ["symboling","carlength","aspiration","carbody","cylindernumber"]
df[categ_names] = df[categ_names].astype('category')
categ_variables = df[categ_names]

categ_variables.describe(include=['category'])

horsepower = df['horsepower']

price = df['price']

def plot_count(x,fig):
    plt.subplot(5,2,fig)
    plt.title(x+' Histogram')
    sns.countplot(x=categ_variables[x],palette=("magma"))
    plt.subplot(5,2,(fig+1))
    plt.title(x+' vs Price')
    sns.boxplot(x=categ_variables[x], y=price, palette=("magma"))
    
plt.figure(figsize=(18,25))

plot_count('symboling', 1)
plot_count('carlength', 3)
plot_count('aspiration', 5)
plot_count('carbody', 7)
plot_count('cylindernumber', 9)

plt.tight_layout()


#sns.catplot(x=current_variable_x, y=current_variable_y, data=categ_variables, kind='box', palette="ch:.25")

# for i in categ_variables:
#      print(categ_variables[i].value_counts())
  
    
```
