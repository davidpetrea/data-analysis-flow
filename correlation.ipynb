```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('./Data/car_data_selection.xlsx')
numeric_variables = df[["price","carwidth","enginesize","stroke","compressionratio","horsepower","peakrpm","boreratio","curbweight","highwaympg","wheelbase","carheight","citympg"]]

#Gets the names of all numeric columns
numeric_variables_names = numeric_variables.columns

plt.figure(figsize=(14,10))
#Pearson Correlation Matrix
corr = numeric_variables.corr(method='pearson', min_periods=1)
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
# Add the mask to the heatmap
sns.heatmap(corr, mask=mask, cmap='YlGnBu', center=0, linewidths=1, annot=True, fmt=".2f")
plt.show()

def scatter_with_ref(variable, fig, ref):
    plt.subplot(5,3,fig)
    plt.scatter(numeric_variables[variable], numeric_variables[ref])
    plt.title(variable+' vs ' + ref)
    plt.ylabel(ref)
    plt.xlabel(variable)

#Set the figure size
plt.figure(figsize=(20,30))
#Set the current reference variable
ref = 'price'

#Loops through all column names of numeric_variables, and calls scatter_with_ref function for each
for index,col in enumerate(numeric_variables_names):
    #skips the currently set reference column
    if(col == ref):
        continue
    scatter_with_ref(col,index,ref)
    
```
